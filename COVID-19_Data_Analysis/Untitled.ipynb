```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('owid-covid-data.csv')

# Display the first few rows to confirm it loaded correctly
df.head()
```

```python
# Get a summary of the DataFrame, including data types and non-null counts
df.info()

# Get descriptive statistics for numerical columns
df.describe()

# Check for the total number of missing values in each column
df.isnull().sum()
```

```python
# Fill missing values in 'total_cases' with 0
df['total_cases'] = df['total_cases'].fillna(0)

# Check the number of missing values again to verify the cleaning
print("Missing values in 'total_cases' after cleaning:")
print(df['total_cases'].isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Filter data for a specific country (e.g., 'India')
india_df = df[df['location'] == 'India']

# Convert the 'date' column to datetime objects for proper plotting
india_df['date'] = pd.to_datetime(india_df['date'])

# Create a line chart of total cases over time
plt.figure(figsize=(12, 7))
plt.plot(india_df['date'], india_df['total_cases'], label='Total Cases')
plt.title('Total COVID-19 Cases in India Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.grid(True)
plt.legend()
plt.show()
```
